```python
f=open("/content/drive/My Drive/Colab Notebooks/s1.txt")
reviews=f.read()
print(reviews)
```

```python
import re
import string
afterHttp = re.sub(r'https?:\/\/.*[\r\n]*', '', reviews)
print(afterHttp)
```

```python
#after #

finalData=re.sub(r'#','',afterHttp)
print(finalData)
```

```python
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize
print(word_tokenize(finalData))
listed=word_tokenize(finalData)
```

```python
from nltk.corpus import stopwords 
from nltk.stem import PorterStemmer

nltk.download('stopwords')
englishWords=stopwords.words('english')
print("english words",englishWords)
```

```python
reviews_clean = []
for word in listed: 
  if (word not in englishWords and word not in string.punctuation):
    reviews_clean.append(word)
print('removed stop words and punctuation:')
print(reviews_clean)
```

```python
stemmer = PorterStemmer()
reviews_stem = []
for word in reviews_clean:
  stem_word = stemmer.stem(word)
  reviews_stem.append(stem_word) 
print('stemmed words:',reviews_stem)
```
