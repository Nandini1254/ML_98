```python
import numpy as np

import matplotlib as plt

from sklearn import datasets #various datasets 

from sklearn import metrics #Check accuracy of model 
```

```python
cancers = datasets.load_breast_cancer()
data=cancers.data
featureNames=cancers.feature_names
targetNames=cancers.target_names
y=cancers.target

print("\nFeature Name: ",featureNames,targetNames)
print("\nFirst 10 rows of data:\n", data[:10])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
   data,y,test_size = 0.4, random_state=1
)
print("training data x \n",X_train)
print("\ntraining data y \n",y_train)
print("\ntesting data x \n",X_test)
print("\ntesting data y \n",y_test)
```
