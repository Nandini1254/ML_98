```python
# input

inputs_p = np.array([[73, 67, 43],
[91, 88, 64],
[87, 134, 58],
[102, 43, 37],
[69, 96, 70]], dtype='float32')

# target

targets_p = np.array([[56],
[81],
[119],
[22],
[103]], dtype='float32')
```

```python
#convert into tenser(like array which will use in toarch)

inputs = torch.from_numpy(inputs_p)
targets = torch.from_numpy(targets_p)


w = torch.randn(1, 3, requires_grad=True)
b = torch.randn(5, requires_grad=True)
print(w)
print(b)
```

```python
#define model --> @ matrix multiplication  
def model(X):
    return X @ w.t() + b

def mse_loss(predictions, targets):
    difference = predictions - targets
    #mean root square 
    return torch.sum(difference * difference)/ difference.numel()

#create model
k=0
for x in inputs:
    preds = model(x)
    print("Prediction is :n",preds)
    print("Target is :g",targets[k])
    print("Loss is: ",mse_loss(preds,targets[k]))
    print()
    k+=1
```

```python
# Compute gradients
epochs = 100
for i in range(epochs):
    # Iterate through training dataloader
    k=0
    for x in inputs:
        # Generate Prediction
        preds = model(x)
        # Get the loss and perform backpropagation
        loss = mse_loss(preds, targets[k])
        k+=1

        # grediat comute
        loss.backward()
        # Let's update the weights
        with torch.no_grad():
            w -= w.grad *1e-6
            b -= b.grad * 1e-6
            # Set the gradients to zero
            w.grad.zero_()
            b.grad.zero_()
    print(f"Epoch {i}/{epochs}: Loss: {loss}")
```
